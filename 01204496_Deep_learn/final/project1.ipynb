```python
import os, sys

pokemon_y = []
fullpath = []
path = "pokemonclassification/PokemonData"
dirs = os.listdir( path )
for files in dirs:
    pokemon_y.append(files)
    files_path = path + "/" + files
    file = os.listdir( files_path )
    img_path = []
    for img in file:
        img_path.append(files_path + "/" + img )
    fullpath.append(img_path) 

# pokemon_y : Keeping the Pokemon name in array
# fullpath  : The path to the each Pokemon image

# print(pokemon_y)
# print(len(pokemon_y))
# print(fullpath)
```

```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread(fullpath[0][1])
plt.imshow(img)
plt.xlabel(pokemon_y[0])
plt.show()
print(img.shape)
```

```python
import cv2
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# convert the image to gray scale : using cvtColor function
```

```python
import numpy as np
from scipy.signal import convolve2d 
Hx = np.array([[ 1,0, -1],[2,0,-2],[1,0,-1]],dtype=np.float32)
Gx = convolve2d(gray,Hx,mode='same')
plt.imshow(Gx,cmap='gray')

Hy = np.array([[ -1,-2, -1],[0,0,0],[1,2,1]],dtype=np.float32)
Gy = convolve2d(gray,Hy,mode='same')
plt.imshow(Gy,cmap='gray')

G = (Gx*Gx + Gy*Gy)**0.5
plt.imshow(G,cmap='gray')
```
