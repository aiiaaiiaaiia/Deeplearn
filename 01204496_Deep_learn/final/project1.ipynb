```python
import numpy as np
from scipy.signal import convolve2d 
import cv2

def SobelFilter(image):
    image_g = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # convert the image to gray scale : using cvtColor function
    
    Hx = np.array([[ 1,0, -1],[2,0,-2],[1,0,-1]],dtype=np.float32)
    img_x = convolve2d(image_g,Hx,mode='same')

    Hy = np.array([[ -1,-2, -1],[0,0,0],[1,2,1]],dtype=np.float32)
    img_y = convolve2d(image_g,Hy,mode='same')

    img = (img_x*img_x + img_y*img_y)**0.5
    return img

img = SobelFilter(img)
plt.xlabel(pokemon_y[1])
plt.imshow(img,cmap='gray')
print(img.shape)
```
