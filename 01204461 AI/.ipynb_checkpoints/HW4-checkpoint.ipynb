```python
#declare variable

#prob
s1_a1_s1 = 0.2
s1_a1_s2 = 0.8
s1_a2_s1 = 0.9
s1_a2_s3 = 0.1

s2_a1_s2 = 0.2
s2_a1_s1 = 0.8
s2_a2_s2 = 0.9
s2_a2_s3 = 0.1

#Rewards
R_s1 = -1
R_s2 = -2
R_s3 = 10

#discount_rate
dr = 0.9

#round
k = 0

#value
V_s1 = 0
V_s2 = 0
V_s3 = 1
```

```python
#round
k = 0

#value
V_s1 = 0.0
V_s2 = 0.0
V_s3 = 1.0

while(k<2):    
    a1_s1 = (s1_a1_s1*(R_s1+(dr*V_s1)))+(s1_a1_s2*(R_s2+(dr*V_s2)))
    a2_s1 = (s1_a2_s1*(R_s1+(dr*V_s1)))+(s1_a2_s3*(R_s3+(dr*V_s3)))
    a1_s2 = (s2_a1_s2*(R_s2+(dr*V_s2)))+(s2_a1_s1*(R_s1+(dr*V_s1)))
    a2_s2 = (s2_a2_s2*(R_s2+(dr*V_s2)))+(s2_a2_s3*(R_s3+(dr*V_s3)))
    
    V_s1 = max( a1_s1 , a2_s1 )
    V_s2 = max( a1_s2 , a2_s2 )
    k += 1
    print("---- k = {} ----".format(k))
    print("V(s1) \t= max( {} , {} ) \t= {}".format(a1_s1 ,a2_s1,V_s1  ))
    print("V(s2) \t= max( {} , {} )\t= {}".format(a1_s2 ,a2_s2,V_s2  ))
    
```

## สำหรับคำนวนคำตอบข้อ 2.2

```python
#round
k = 0

#value
V_s1 = 0.0
V_s2 = 0.0
V_s3 = 1.0

while(k<100):
    a1_s1 = (s1_a1_s1*(R_s1+(dr*V_s1)))+(s1_a1_s2*(R_s2+(dr*V_s2)))
    a2_s1 = (s1_a2_s1*(R_s1+(dr*V_s1)))+(s1_a2_s3*(R_s3+(dr*V_s3)))
    a1_s2 = (s2_a1_s2*(R_s2+(dr*V_s2)))+(s2_a1_s1*(R_s1+(dr*V_s1)))
    a2_s2 = (s2_a2_s2*(R_s2+(dr*V_s2)))+(s2_a2_s3*(R_s3+(dr*V_s3)))
    
    V_s1 = max( a1_s1 , a2_s1 )
    V_s2 = max( a1_s2 , a2_s2 )
    k +=1 
    print(" k = {} \t|\tV(s1) = {} \t|\tV(s2) = {} ".format(k, V_s1, V_s2))
    
```

## สำหรับคำนวนคำตอบข้อ 2.3

```python
#round
k = 0

#value
V_s1 = 0.0
V_s2 = 0.0
V_s3 = 1.0

while(k<100):
    a1_s1 = (s1_a1_s1*(R_s1+(dr*V_s1)))+(s1_a1_s2*(R_s2+(dr*V_s2)))
    a2_s1 = (s1_a2_s1*(R_s1+(dr*V_s1)))+(s1_a2_s3*(R_s3+(dr*V_s3)))
    a1_s2 = (s2_a1_s2*(R_s2+(dr*V_s2)))+(s2_a1_s1*(R_s1+(dr*V_s1)))
    a2_s2 = (s2_a2_s2*(R_s2+(dr*V_s2)))+(s2_a2_s3*(R_s3+(dr*V_s3)))
    
    if(a1_s1 >= a2_s1):
        V_s1 = a1_s1
        path_s1 = "action 1"
    else:
        V_s1 = a2_s1
        path_s1 = "action 2"

    if(a1_s2 >= a2_s2):
        V_s2 = a1_s2
        path_s2 = "action 1"
    else:
        V_s2 = a2_s2
        path_s2 = "action 2"

    k +=1 
    print(" k = {} \t|\tV(s1) = {}  \t|\tV(s2) = {}   \t|\t state 1 is take  :  {}  \t|\t state 2 is take  :  {} \t|".format(k, V_s1, V_s2, path_s1, path_s2))
    
```

## สำหรับคำนวนคำตอบข้อ 3

```python
import numpy as np
import matplotlib.pyplot as plt 
X = np.arange(0,1.0002,0.0001)
Y1=[]
Y2=[]

for x in X:
  y1=50*x-(pow(x, 2)*(1-pow(x, 100))/(1-x))
  Y1.append(y1)

for x in X:
  y1=-50*x+(pow(x, 2)*(1-pow(x, 100))/(1-x))
  Y2.append(y1)

plt.plot(X, Y1, label = "Up",  color='red') 
plt.plot(X, Y2, label = "down", color='blue') 
  
plt.xlabel('discount value') 
plt.ylabel('utility') 
plt.title('The utility of each action as a function of discount value') 
  
# plt.legend()
plt.show()
```
