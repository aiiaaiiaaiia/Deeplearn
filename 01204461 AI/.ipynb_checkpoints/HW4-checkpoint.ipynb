### compute value iteration

```python
#declare variable

#prob
s1_a1_s1 = 0.2
s1_a1_s2 = 0.8
s1_a2_s1 = 0.9
s1_a2_s3 = 0.1

s2_a1_s2 = 0.2
s2_a1_s1 = 0.8
s2_a2_s2 = 0.9
s2_a2_s3 = 0.1

#Rewards
R_s1 = -1
R_s2 = -2
R_s3 = 10

#discount_rate
dr = 0.9

#round
k = 0

#value
V_s1 = 0
V_s2 = 0
V_s3 = 1
```

```python
while(k<100):
    V_s1 = max( (s1_a1_s1*(R_s1+(dr*V_s1)))+(s1_a1_s2*(R_s2+(dr*V_s2))) , (s1_a2_s1*(R_s1+(dr*V_s1)))+(s1_a2_s3*(R_s3+(dr*V_s3))) )
    V_s2 = max( (s2_a1_s2*(R_s2+(dr*V_s2)))+(s2_a1_s1*(R_s1+(dr*V_s1))) , (s2_a2_s2*(R_s2+(dr*V_s2)))+(s2_a2_s3*(R_s3+(dr*V_s3))) )
    k +=1 
    print(" k = {} \t|\tV(s1) = {} \t|\tV(s2) = {} ".format(k, V_s1, V_s2))
    
```
